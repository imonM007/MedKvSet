```python
import os
import cv2
import numpy as np
import pandas as pd
from glob import glob
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader, Subset
from sklearn.model_selection import KFold
import matplotlib.pyplot as plt

# ==========================================================
# 1. DATASET CLASS (256x256 Resolution)
# ==========================================================
class SegNetDataset(Dataset):
    def __init__(self, image_dir, mask_dir):
        # Finds .jpg, .jpeg, and .png files (case-insensitive)
        self.image_paths = sorted(glob(os.path.join(image_dir, "*.jpg")) + 
                                  glob(os.path.join(image_dir, "*.png")) + 
                                  glob(os.path.join(image_dir, "*.jpeg")))
        self.mask_paths = sorted(glob(os.path.join(mask_dir, "*.png")) + 
                                 glob(os.path.join(mask_dir, "*.jpg")) + 
                                 glob(os.path.join(mask_dir, "*.jpeg")))
        
        if len(self.image_paths) == 0:
            raise ValueError(f"No images found. Check path: {image_dir}")

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        image = cv2.imread(self.image_paths[idx], cv2.IMREAD_COLOR)
        mask = cv2.imread(self.mask_paths[idx], cv2.IMREAD_GRAYSCALE)
        
        image = cv2.resize(image, (256, 256))
        mask = cv2.resize(mask, (256, 256))
        
        image = image.transpose((2, 0, 1)) / 255.0  # To CHW
        mask = mask[np.newaxis, ...] / 255.0        # To CHW
        return torch.tensor(image, dtype=torch.float32), torch.tensor(mask, dtype=torch.float32)

# ==========================================================
# 2. TRUE SEGNET ARCHITECTURE (CPU-Optimized Slim Version)
# ==========================================================
class SlimSegNet(nn.Module):
    def __init__(self):
        super(SlimSegNet, self).__init__()
        def conv_block(in_c, out_c):
            return nn.Sequential(
                nn.Conv2d(in_c, out_c, kernel_size=3, padding=1),
                nn.BatchNorm2d(out_c),
                nn.ReLU(inplace=True)
            )
        # Reduced channel depth (16-128) for CPU speed
        self.enc1 = conv_block(3, 16); self.enc2 = conv_block(16, 32)
        self.enc3 = conv_block(32, 64); self.enc4 = conv_block(64, 128)
        self.pool = nn.MaxPool2d(2, stride=2, return_indices=True)
        self.unpool = nn.MaxUnpool2d(2, stride=2)
        self.dec4 = conv_block(128, 64); self.dec3 = conv_block(64, 32)
        self.dec2 = conv_block(32, 16); self.dec1 = nn.Conv2d(16, 1, kernel_size=3, padding=1)

    def forward(self, x):
        # Encoder
        x1 = self.enc1(x); p1, idx1 = self.pool(x1)
        x2 = self.enc2(p1); p2, idx2 = self.pool(x2)
        x3 = self.enc3(p2); p3, idx3 = self.pool(x3)
        x4 = self.enc4(p3); p4, idx4 = self.pool(x4)

        # Decoder using Pooling Indices
        d4 = self.dec4(self.unpool(p4, idx4, output_size=x4.size()))
        d3 = self.dec3(self.unpool(d4, idx3, output_size=x3.size()))
        d2 = self.dec2(self.unpool(d3, idx2, output_size=x2.size()))
        d1 = torch.sigmoid(self.dec1(self.unpool(d2, idx1, output_size=x1.size())))
        return d1

# ==========================================================
# 3. COMPREHENSIVE METRICS & VISUALIZATION
# ==========================================================
def get_metrics(pred, target, threshold=0.5):
    pred, target = (pred > threshold).float(), (target > 0.5).float()
    tp, fp, fn, tn = (pred * target).sum().item(), (pred * (1-target)).sum().item(), \
                     ((1-pred) * target).sum().item(), ((1-pred) * (1-target)).sum().item()
    eps = 1e-7
    return {
        "accuracy": (tp + tn) / (tp + tn + fp + fn + eps),
        "precision": tp / (tp + fp + eps), "recall": tp / (tp + fn + eps),
        "f1": 2*tp / (2*tp + fp + fn + eps), "iou": tp / (tp + fp + fn + eps), 
        "dice": 2*tp / (2*tp + fp + fn + eps)
    }

def show_predictions(model, dataset, title="Progress Update"):
    model.eval()
    device = next(model.parameters()).device
    indices = np.random.choice(len(dataset), 3, replace=False)
    plt.figure(figsize=(12, 8))
    for i, idx in enumerate(indices):
        img, mask = dataset[idx]
        with torch.no_grad():
            pred = model(img.unsqueeze(0).to(device))
            pred_mask = (pred.squeeze().cpu().numpy() > 0.5).astype(np.uint8)
        
        plt.subplot(3, 3, i*3 + 1); plt.imshow(img.permute(1,2,0)); plt.title("Original Image"); plt.axis('off')
        plt.subplot(3, 3, i*3 + 2); plt.imshow(mask.squeeze(), cmap='gray'); plt.title("Ground Truth"); plt.axis('off')
        plt.subplot(3, 3, i*3 + 3); plt.imshow(pred_mask, cmap='gray'); plt.title(f"{title} Prediction"); plt.axis('off')
    plt.tight_layout(); plt.show()

# ==========================================================
# 4. MAIN TRAINING LOOP (K-Fold + 100 Epochs + Early Stopping)
# ==========================================================
def train_kfold():
    # Update these paths to match your local folders
    image_dir = r'E:\BIOSE\Dataset\image'
    mask_dir = r'E:\BIOSE\Dataset\label'
    device = "cuda" if torch.cuda.is_available() else "cpu"
    
    k_folds = 5
    max_epochs = 100
    patience = 7  # Stops if val_loss doesn't improve for 7 epochs
    
    dataset = SegNetDataset(image_dir, mask_dir)
    kfold = KFold(n_splits=k_folds, shuffle=True, random_state=42)
    history_log = []

    for fold, (train_ids, val_ids) in enumerate(kfold.split(dataset)):
        print(f"\n" + "="*30 + f" FOLD {fold+1}/{k_folds} " + "="*30)
        train_sub, val_sub = Subset(dataset, train_ids), Subset(dataset, val_ids)
        train_loader = DataLoader(train_sub, batch_size=4, shuffle=True)
        val_loader = DataLoader(val_sub, batch_size=4, shuffle=False)
        
        model = SlimSegNet().to(device)
        optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)
        criterion = nn.BCELoss()
        best_v_loss = float('inf'); no_improve = 0

        for epoch in range(max_epochs):
            # Training Phase
            model.train()
            t_loss, t_m = 0, {k: 0 for k in ["accuracy", "dice", "iou", "precision", "recall", "f1"]}
            for i, (imgs, msks) in enumerate(train_loader):
                imgs, msks = imgs.to(device), msks.to(device)
                out = model(imgs); loss = criterion(out, msks)
                optimizer.zero_grad(); loss.backward(); optimizer.step()
                t_loss += loss.item()
                m = get_metrics(out, msks); [t_m.update({k: t_m[k] + m[k]}) for k in t_m]
                if i % 100 == 0: print(f"Fold {fold+1} | Ep {epoch+1} | Batch {i}/{len(train_loader)}...")

            # Validation Phase
            model.eval()
            v_loss, v_m = 0, {k: 0 for k in ["accuracy", "dice", "iou", "precision", "recall", "f1"]}
            with torch.no_grad():
                for imgs, msks in val_loader:
                    imgs, msks = imgs.to(device), msks.to(device)
                    out = model(imgs); v_loss += criterion(out, msks).item()
                    m = get_metrics(out, msks); [v_m.update({k: v_m[k] + m[k]}) for k in v_m]

            nt, nv = len(train_loader), len(val_loader)
            v_avg_loss = v_loss / nv
            print(f"--- F{fold+1} Ep {epoch+1} Results: Val Loss: {v_avg_loss:.4f} | Val Dice: {v_m['dice']/nv:.4f} ---")
            
            # Logging all metrics
            log = {"fold": fold+1, "epoch": epoch+1, "train_loss": t_loss/nt, "val_loss": v_avg_loss}
            log.update({f"train_{k}": v/nt for k, v in t_m.items()})
            log.update({f"val_{k}": v/nv for k, v in v_m.items()})
            history_log.append(log)

            # Early Stopping Check
            if v_avg_loss < best_v_loss:
                best_v_loss = v_avg_loss; no_improve = 0
            else:
                no_improve += 1
                if no_improve >= patience:
                    print(f"Early stopping triggered for Fold {fold+1}. Moving to next fold.")
                    break

        # Prediction output after each fold
        show_predictions(model, val_sub, title=f"Fold {fold+1}")

    # Save to CSV
    pd.DataFrame(history_log).to_csv("segnet_1800.csv", index=False)
    print("\nTraining Complete. Full metrics saved to segnet_1800.csv")
    return pd.DataFrame(history_log), model, val_sub

if __name__ == "__main__":
    final_df, last_model, last_val_ds = train_kfold()
    print("\n" + "="*20 + " FINAL PERFORMANCE SUMMARY " + "="*20)
    show_predictions(last_model, last_val_ds, title="Final Model")
```
